```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete.head()
#Calculate the total number of schools
total_schools = school_data_complete["school_name"].nunique()

#Calculate the total number of students
total_students = school_data_complete["Student ID"].count()
total_students

# Calculate the total budget / first created a group by and used Max since the repeating rows would have the same value
#t Then summed up that column
total_budget = school_data_complete.groupby(["school_name"]).max()
total_budget_sum = total_budget["budget"].sum().astype(float)

#Calculate the average math score
average_math = round(school_data_complete["math_score"].mean(),2)

#Calculate the average reading score
average_read = round(school_data_complete["reading_score"].mean(),2)

#Calculate the percentage of students with a passing math score (70 or greater)
passing_math_count = school_data_complete[school_data_complete["math_score"]>69].count()
passing_math_avg = round(passing_math_count/total_students*100,2)

#Calculate the percentage of students with a passing reading score (70 or greater)
passing_read_count = school_data_complete[school_data_complete["reading_score"]>69].count()
passing_read_avg = round(passing_read_count/total_students*100,2)

#Calculate the percentage of students who passed math **and** reading (% Overall Passing)
pass_read_math_count = school_data_complete[(school_data_complete["reading_score"]>69)& (school_data_complete["math_score"]>69)].count()
pass_read_math_avg = round(pass_read_math_count/total_students*100,2)


# Creating a summary DataFrame using above values
summary_df = pd.DataFrame([{"Total Schools": total_schools,
                            "Total Students": total_students,
                            "Total Budget": total_budget_sum,
                            "Average Math Score": average_math,
                            "Average Reading Score": average_read,
                            "% Passing Math": passing_math_avg.max(),
                            "% Passing Reading": passing_read_avg.max(),
                           "% Overall Passing": pass_read_math_avg.max(),}])

summary_df['Total Budget'] = summary_df['Total Budget'].map('${:,.2f}'.format)
summary_df['Total Students'] = summary_df['Total Students'].map('{:,}'.format)

summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Grouped our complete data frame by school name 
grouped_school = school_data_complete.groupby(['school_name'])

# Calculate the total student for each school
total_student = grouped_school.size()

# Get the school type for each school
school_type = grouped_school['type'].max()

# Calculate the total budget for each school
total_budget = grouped_school['budget'].max()

# Calculate the budget per student for each school
budget_per_student = total_budget/total_student

# Calculate the average math score for each school
math_score_avg = grouped_school['math_score'].mean()

# Calculate the average reading score for each school
reading_score_avg = grouped_school['reading_score'].mean()

# Calculate the percentange of passing math score for each school
math_passing_grp = school_data_complete[school_data_complete['math_score']>69].groupby(['school_name']).size()
percent_passing_math = (math_passing_grp/total_student)*100

# Calculate the percentange of passing math score for each school
reading_passing_grp = school_data_complete[school_data_complete['reading_score']>69].groupby(['school_name']).size()
percent_passing_reading = (reading_passing_grp/total_student)*100

# Calculate the overall passing score for each school
both_passing_grp = school_data_complete[(school_data_complete['reading_score']>69)&(school_data_complete['math_score']>69)].groupby(['school_name']).size()
percent_overall_passing = both_passing_grp/total_student*100

# Create a dataframe to hold the above results
school_summary = pd.DataFrame({
    'School Type': school_type,
    'Total Students':total_student,
    'Total School Budget': total_budget.map('${:,.2f}'.format),
    'Per Student Budget': budget_per_student,
    'Average Math Score': round(math_score_avg,2),
    'Average Reading Score': round(reading_score_avg,2),
    '% Passing Math': round(percent_passing_math,2),
    '% Passing Reading': round(percent_passing_reading,2),
    '% Overall Passing Rate': round(percent_overall_passing,2),
})

school_summary
```
